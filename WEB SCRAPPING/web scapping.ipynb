# 1.get the data

```python
import requests as r
url="https://www.typing.com"
r=r.get(url)
data=r.content#here we take content it takes as a bite and if we take .text it takes as a string or text
print(data)
```

# 2.parse the data

here parshing means to converts something into something so here we takes and data in string or binary format so we have to convert that in html code so wil use html.parser apart from that there are many parser and this all stuff is in butifulsoup (bs4)

```python
from bs4 import BeautifulSoup
obj=BeautifulSoup(data,'html.parser')
obj
#now it is a code in html not a string or bytes
#if you want to see it in some structured way then you use 
print(obj.prettify)
```

```python
# with the use of obj will scrape the data with some funtions like 
obj.find_all("li")
obj.find("ul")["class"]
```

```python
# we want to get all link from data 
a_tag=obj.find_all("a")
link=[]
for i in a_tag:
    if i!="#":
        link.append(i.get("href"))
print(link)
```
